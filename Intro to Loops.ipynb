```python
#case 1 if ,elif
total3=0
total4=0
for i in range(1,100):
    if i%3 == 0:   #3,6,9........
       total3 +=i
    elif i%5 == 0:    #5,10.....
        total4 +=i
print(total3)  
print(total4) 
print(total3 + total4)
```

```python
#case 2 if
total3=0
 
for i in range(1,100):
    if i%3 == 0 or i%5 == 0: 
       total3 +=i   
print(total3)  
```

```python
#case 3 if ,elif
total4 =0
for i in range(1,100):
    if i%3 == 0:   #3,6,9........
       total4 +=i
    elif i%5 == 0:    #5,10.....
        total4 +=i
print(total4)  
```
