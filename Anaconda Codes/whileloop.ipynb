```python
# case 1
# to print sum of number 0,1,2,3,4 (<5) == 10 
total=0 
j=1
while j<5:
    total += j
    j+=1
print(total)    
```

```python
#case 2
# to print sum of all positive number in arraylist using while loop == 17
given_list=[5,4,4,3,1,-2,-3,-5] 
total2=0
i=0
while given_list[i]>0:
    total2 += given_list[i]
    i += 1
print(total2)
```

```python
#case 3
# to print sum of all +ve no. using for,if loop == 28/iff if loop not used answer == 18 (+,-)
given_list2=[6,5,4,3,9,1,-2,-3,-5]
total3=0
for element in given_list2:
    if element <= 0:
        break
    total3 += element
print(total3)
```

```python
#case 4 
# given_list2=[6,5,4,3,9,1,-2,-3,-5] using diff way of while loop
total3=0
i=0
while True:
    total3 += given_list2[i]
    i += 1
    if given_list2[i] <= 0:
        break
print(total3)        
```

```python
# to print =ve no. sum ==-18
given_list3=[6,6,5,4,3,9,1,-1,-2,-3,-5,-7]
total4 =0
j = len(given_list3) - 1
while given_list3[j] < 0:
    total4 += given_list3[j]
    j -= 1
print(total4)
```
