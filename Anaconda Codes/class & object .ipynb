```python
class Robot:
    def __init__(self,name,color,weight):
        self.name = name 
        self.color = color
        self.weight = weight
        
    def introduce_self(self):
        print("My name is " + self.name) #this in java
```

```python
class Person:
    def __init__(self, n, p, i):
        self.name = n 
        self.personality = p
        self.is_sitting = i
    
    def sit_down(self):
        self.is_sitting = True

    def stand_up(self):
        self.is_sitting = False
```

```python
p1 = Person("Allice","aggressive", False)
p2 = Person("Backer","talkactive",True)
```

```python
#p1 owns r2 
p1.robot_owned =r2
p2.robot_owned =r1
#p2 owns r1
```

```python
p1.robot_owned.introduce_self()
```
