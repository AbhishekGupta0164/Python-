```python
given_list =[1,1,2,4,2]
```

```python
new_set1 = set()
for x in given_list:
    new_set1.add(x)
print(new_set1)    
```

```python
#toprint a new list of unique element
new_list1 = list()
for x in new_set1:
    new_list1.append(x)
print(new_list1)    
```

```python
b = set()
b.add('apple')
b.add('banana')
b.add('car')
b.add(1)
print(b)
```

```python
given_list2 =[2,3,4,4,9,8]
```

```python
# to print sum of unique element 
new_set2 =set()
for x in given_list2:
    new_set2.add(x)
total = 0
for x in new_set2:
    total += x
print(total)
```
